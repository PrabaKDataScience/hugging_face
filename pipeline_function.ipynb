```python
from transformers import pipeline

def do_sentiment_analysis(sentence:str):

    # Sentiment analysis
    sent_analysis_pp = pipeline(task = "sentiment-analysis", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
    score = sent_analysis_pp(sentence)
    return score



# Sentiment analsysis
sentiment = do_sentiment_analysis(sentence="The product price is high but quality isnt that much")
sentiment


```

```python
def classify_within_options(sentence , categories):
    classifier  = pipeline(task='zero-shot-classification',
                               model="facebook/bart-large-mnli")

    return classifier(sentence, categories)

sequence_to_classify = "one day I will make cake"
candidate_labels = ['travel', 'cooking', 'dancing']
classified = classify_within_options(sentence=sequence_to_classify,categories=candidate_labels)
classified
```
